```python
import numpy as np
import matplotlib.pyplot as plt

# -----------------------------
# Step 1: Generate Noisy 2D Data
# -----------------------------

cluster1 = np.random.randn(100, 2) + [3, 3]
cluster2 = np.random.randn(100, 2) + [-3, -3]
cluster3 = np.random.randn(100, 2) + [3, -3]

data = np.vstack((cluster1, cluster2, cluster3))
# -----------------------------
# Step 2: Initialize Network
# -----------------------------
num_neurons = 3
weights = np.random.randn(num_neurons, 2)

learning_rate = 0.1
epochs = 50

# -----------------------------
# Step 3: Training Process
# -----------------------------
for epoch in range(epochs):
    for x in data:

        # Compute distance
        distances = np.linalg.norm(weights - x, axis=1)

        # Winner neuron
        winner = np.argmin(distances)

        # Update weights
        weights[winner] += learning_rate * (x - weights[winner])

    # Reduce learning rate
    learning_rate *= 0.95

# -----------------------------
# Step 4: Results Visualization
# -----------------------------
plt.figure()
plt.scatter(data[:, 0], data[:, 1], alpha=0.4, label="Noisy Data")
plt.scatter(weights[:, 0], weights[:, 1], s=200, label="Final Neurons")
plt.title("2-Layer Competitive Learning Network")
plt.legend()
plt.show()

print("Final Weights (Neuron Positions):")
print(weights)
```
